```python
import random
import os

move_list = ['papel', 'pedra', 'tesoura']
player_count = 0
computer_count = 0

print('========================')
print('Bem vindo ao jogo Pedra, Papel e Tesoura')


def main_print():
    print('==============================')
    print('\nPlacar')
    print(f'Você: {player_count}')
    print('\n')
    print('Escolha seu lance:')
    print('0 - papel | 1 - pedra | 2 - tesoura')
    

def select_move():
    return random.choice(move_list)

def get_player_move():
    while True:
        try:
            player_move = int(input())
            if player_move not in [0,1,2]:
                raise
            return move_list[player_move]
        except Exception as e:
            print(e)

def select_winner(p_movie, c_movie):
    if p_movie == 'papel':
        if c_movie == 'pedra':
            player_count += 1
            return 'p'
        
        elif c_movie == 'tesoura':
            computer_count += 1
            return 'c'
        
        else:
            return 'd'
        
    if p_movie == 'pedra':
        if c_movie == 'tesoura':
            player_count += 1
            return 'p'
        
        elif c_movie == 'papel':
            computer_count += 1
            return 'c'
        
        else:
            return 'd'
        
    if p_movie == 'tesoura':
        if c_movie == 'papel':
            player_count += 1
            return 'p'
        
        elif c_movie == 'pedra':
            computer_count += 1
            return 'c'
        
        else:
            return 'd'
        
again = 1
while again == 1:
    main_print()
    player_movie = get_player_move()
    computer_movie = select_move()
    winner = select_winner(player_movie, computer_movie)
    
    print('')
    print('==========================')
    print(f'Sua jogada: {player_movie.upper()}')
    print(f'Jogada do computador: {computer_movie.upper()}')
    
    if winner == 'p':
        print('Você venceu!')
    elif winner == 'c':
        print('Você perdeu!')
    else:
        print('Empate!')
    print('===========================')
    
    while True:
        print('Jogar novamente? 0 - sim | 1 - não')
        next = int(input())
        if next == 0:
            break
        elif next == 1:
            again = 0
            break
    
    os.system('clear')
```
